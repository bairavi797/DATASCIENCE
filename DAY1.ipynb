```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(20)
data = {
    "x" : np.arange(1,8),
    "y" : np.random.randint(20,30,size=7)
}
df = pd.DataFrame(data)
print(df)
plt.plot(df["x"],df["y"],marker='*',linestyle='--',color = 'purple')
plt.xlabel("days")
plt.ylabel("sales")
plt.title("LINE PLOT")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(20)
data = {
    "x" : np.arange(1,8),
    "y" : np.random.randint(20,30,size=7)
}
df = pd.DataFrame(data)
plt.barh(df["x"],df["y"],color = 'red')
plt.xlabel("days")
plt.ylabel("sales")
plt.title("HORIZONTAL BAR")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(20)
data = {
    "x" : np.arange(1,8),
    "y" : np.random.randint(20,30,size=7)
}
df = pd.DataFrame(data)
plt.hist(df["y"],bins=5,color = 'green')
plt.xlabel("days")
plt.ylabel("sales")
plt.title("Sales Analysis")
plt.grid(False)
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ["A", "B", "C", "D"]
sizes = [30, 25, 20, 25]

plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.title("Simple Pie Chart")
plt.show()
```
